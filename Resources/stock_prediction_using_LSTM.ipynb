# 4. Data Preprocessing

```python
df.shape
```

# spliting data into testing and training datasets

```python
data_training= pd.DataFrame(df.Close[0:int(len(df)*0.7)])
data_testing= pd.DataFrame(df.Close[int(len(df)*0.7):int(len(df))])

print(data_training)
print(data_testing)
```

```python
data_training
```

```python
data_testing
```

## Scaling Down The Data Between 0 and 1

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
type(data_training_array)
```

```python
data_training_array.shape
```

# 5. Data Sequence Creation

```python
# 12,13,14,15,16  predics 14 -- past data are x_train and 14 is y_train
# if we want next prediction it takes y_trians as lastdata and remove first element become x_train
# 13,14,15,16,14    --- 18

x_train=[]
y_train=[]

for i in range(100,data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i,0])

x_train,y_train=np.array(x_train),np.array(y_train)
```

```python
y_train.shape
```

# 6. Model Building

```python
import tensorflow as tf
print(tf.__version__)
```

```python
# ML Model
from keras.models import Sequential
from keras.layers import Dense,LSTM,Input
```

```python
#model init
model = Sequential()

model.add(Input(shape=(100,1)))
model.add(LSTM(units=128,activation='tanh',return_sequences=True))
model.add(LSTM(units=64))
model.add(Dense(25))
model.add(Dense(1))
```

# 7. Model Training

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=50)
```

```python
model.summary()
```

```python
# Save the trained model
model.save('stock_prediction_model.keras')
```

# 9. Make Prediction

```python
y_predicted = model.predict(x_test)
```

# 10.Model Evaluation

```python
# mean squred error(MSE)
# low means model predicted well
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predicted)
print(f"Mean Squared Error (MSE): {mse}")
```

```python
# Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
# R-Squared
r2 = r2_score(y_test, y_predicted)
print(f"R-Squared: {r2}")
```
